### Question 1) Regular plots [5 points]

Starting from the code below, try to reproduce the graphic below taking care of filled areas.

```python
import numpy as np
import matplotlib.pyplot as plt

n = 256
X = np.linspace(-np.pi,np.pi,n,endpoint=True)
Y = np.sin(2*X)

plt.plot (X, Y+1, color='blue', alpha=1.00)
plt.plot (X, Y-1, color='blue', alpha=1.00)
plt.show()
```

```python
# YOUR CODE HERE
n = 256
X = np.linspace(-np.pi,np.pi,n,endpoint=True)
Y = np.sin(2*X)

fig1, ax1 = plt.subplots()
ax1.fill(X, Y+1, color='lightblue')
ax1.plot(X, Y+1, color='blue', alpha=1.00)

ax1.fill_between(X, Y-1, -1, where=(Y-1 > -1), color='lightblue')
ax1.fill_between(X, Y-1, -1, where=(Y-1 <= -1), color='red')
ax1.plot(X, Y-1, color='blue', alpha=1.00)
```

### Question 4) Contour plots [5 points]

Starting from the code below, try to reproduce the graphic below taking care of the colormap.

```python
import numpy as np
import matplotlib.pyplot as plt

def f(x,y): return (1-x/2+x**5+y**3)*np.exp(-x**2-y**2)

n = 256
x = np.linspace(-3,3,n)
y = np.linspace(-3,3,n)
X,Y = np.meshgrid(x,y)

plt.contourf(X, Y, f(X,Y), 8, alpha=.75, cmap='jet')
C = plt.contour(X, Y, f(X,Y), 8, colors='black', linewidth=.5)
plt.show()
```

```python
# YOUR CODE HERE
def f(x,y): return (1-x/2+x**5+y**3)*np.exp(-x**2-y**2)

n = 256
x = np.linspace(-3,3,n)
y = np.linspace(-3,3,n)
X,Y = np.meshgrid(x,y)

plt.contourf(X, Y, f(X,Y), 7, alpha=.75, cmap='hot')
C = plt.contour(X, Y, f(X,Y), 7, colors='black', linewidth=.25)
plt.clabel(C, inline=True, fmt='%1.3f', fontsize=9)

plt.xticks(())
plt.yticks(())
plt.show()
```

### Question 6) Pie charts [5 points]

<img src="pie_charts.png">

```python
# YOUR CODE HERE

n = 20 - 1
Z = np.ones(n)
Z[-1] = 2

colors = np.linspace(0,1,20)
color_hex = [plt.cm.grey(c) for c in colors]

explode = np.zeros(n) + 0.075
explode[n - 1] = 0.1

plt.pie(Z, colors=color_hex, 
        wedgeprops={'linewidth': 1, 'edgecolor': 'black'},
        startangle=0, explode=explode)
plt.show()
```
